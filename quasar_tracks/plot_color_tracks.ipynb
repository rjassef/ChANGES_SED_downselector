```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
def read_davenport():

    #Read the stellar locus fits form Davenport et al. (2014). We will use the following colors for the analysis: 
    #g-i, r-i, r-z, z-W1, r-W1
    dv14 = Table.read("../Davenport14/table1.txt",format='ascii')
    dv14_nir = Table.read("../Davenport14/table2.txt",format='ascii')
    sl_tab = Table()
    sl_tab['g-i']  = dv14['g-i']
    sl_tab['r-i']  = dv14['r-i']
    sl_tab['r-z']  = dv14['r-i'] + dv14['i-z']
    sl_tab['z-w1'] = dv14['z-J'] + dv14['J-H'] + dv14['H-Ks'] + dv14_nir['Ks-W1']
    sl_tab['r-w1'] = dv14['r-i'] + dv14['i-z'] + dv14['z-J'] + dv14['J-H'] + dv14['H-Ks'] + dv14_nir['Ks-W1']

    return sl_tab

sl_tab = read_davenport()
```

```python
colors = Table.read("quasar_tracks.dat", format='ascii')
```

```python
plt.plot(colors['r-i'], colors['r-w1'], '-b')
plt.plot(sl_tab['r-i'], sl_tab['r-w1'], '-r')
plt.xlim([-1.0, 3.0])
plt.ylim([1.5, 7.5])
plt.show()
```

```python
plt.plot(colors['g-i'], colors['z-w1'], '-b')
plt.plot(sl_tab['g-i'], sl_tab['z-w1'], '-r')
plt.xlim([-3.0, 10.0])
plt.ylim([1.0, 7.5])
plt.show()
```

```python
plt.plot(colors['g-i'], colors['r-z'], '-b')
plt.plot(sl_tab['g-i'], sl_tab['r-z'], '-r')
plt.xlim([-3.0, 8.0])
plt.ylim([-1.5, 4.0])
plt.show()
```

```python
plt.plot(colors['r-i'], colors['z-w1'], '-b')
plt.plot(sl_tab['r-i'], sl_tab['z-w1'], '-r')
plt.xlim([-1.5, 2.5])
plt.ylim([1.0, 7.5])
plt.show()
```
